## Exploratory Data Analysis (EDA)

### Explore ...

Analysis of Sales Performance of Each Vendor in the Last Year

```python
# Filter data for the past year
last_year = orders_df['order_purchase_timestamp'].max() - pd.DateOffset(years=1)
orders_last_year = orders_df[orders_df['order_purchase_timestamp'] >= last_year]
```

```python
# Merge orders and order_items datasets to get details of orders shipped
orders_items_last_year = pd.merge(orders_last_year, order_items_df, on='order_id')
```

```python
# Merge with sellers to get seller information
sales_performance = pd.merge(orders_items_last_year, sellers_df, on='seller_id')
```

```python
# Calculate total sales (revenue) and order quantity per seller
sales_performance_summary = sales_performance.groupby(['seller_id', 'seller_city', 'seller_state']).agg(
    total_sales=('price', 'sum'),
    total_orders=('order_id', 'nunique')
).reset_index()
```

```python
# Sort by highest total sales
top_sellers = sales_performance_summary.sort_values(by='total_sales', ascending=False)
```

```python
# Show 5 sellers with the highest total sales
print("Top 5 Sellers by Total Sales:")
print(top_sellers.head())
```

Analysis of Differences in Shopping Patterns between Old and New Customers and Their Contribution to Total Revenue

```python
# Combine `orders_df` with `customers_df` to get customer and order data
customer_orders = pd.merge(orders_df, customers_df, on='customer_id')
```

```python
# Identify existing and new customers
# New customers are those who have only 1 order
customer_order_count = customer_orders['customer_unique_id'].value_counts().reset_index()
customer_order_count.columns = ['customer_unique_id', 'total_orders']
customer_orders = pd.merge(customer_orders, customer_order_count, on='customer_unique_id')
```

```python
# Categorize customers into 'old' and 'new'
customer_orders['customer_type'] = customer_orders['total_orders'].apply(lambda x: 'Pelanggan Baru' if x == 1 else 'Pelanggan Lama')
```

```python
# Combine with `order_items_df` to get details of items ordered and revenue generated
customer_order_items = pd.merge(customer_orders, order_items_df, on='order_id')
```

```python
# Calculate total revenue per customer type
revenue_per_customer_type = customer_order_items.groupby('customer_type').agg(
    total_revenue=('price', 'sum'),
    total_orders=('order_id', 'nunique')
).reset_index()
```

```python
# Calculate the contribution of each to total revenue
total_revenue = revenue_per_customer_type['total_revenue'].sum()
revenue_per_customer_type['revenue_contribution'] = revenue_per_customer_type['total_revenue'] / total_revenue * 100
```

```python
# Show results
print("Revenue per Customer Type:")
print(revenue_per_customer_type)
```

**Insight:**

**Analysis of Sales Performance of Each Vendor in the Last Year**:
- Based on the data exploration, I got a new insight that we can see some sellers with the highest total sales in the past year. Sellers with the highest revenue indicate that they have popular products or effective sales strategies.
- In addition to total sales, the number of orders per seller provides information on how often orders are placed by customers to that seller. Sellers with many orders and high total sales may indicate that they have high customer loyalty.
- It is known that it is the seller's city or region that generates the highest sales. We can find out the areas that may have high demand for products.
- From low sales, we can further explore whether there are certain constraints that cause their performance to be less than optimal. Whether the constraints are related to logistics, pricing, or inappropriate marketing strategies.

**Analysis of Differences in Shopping Patterns between Old and New Customers and Their Contribution to Total Revenue**:
- From the data analysis, the revenue contribution from new customers is higher than that of existing customers. This indicates that new customers tend to make the first purchase with a larger transaction value, perhaps due to special incentives or strong initial interest in the product.
- Higher revenue from new customers shows the great potential of customer acquisition strategies. However, it is important to encourage them to become loyal customers and make repeat purchases. If new customers only make one purchase, it will be difficult for businesses to maintain long-term revenue.
- Existing customers who have lower revenue contribution could indicate that they may have lost interest or are not interested in the latest offerings. There needs to be further analysis to understand what factors lead to low revenue contribution from existing customers.
- Since the contribution of new customers is dominant, focusing on retention strategies is essential. Retaining new customers to make a second and subsequent purchase will help increase revenue and loyalty. Otherwise, businesses will need to constantly seek new customers to maintain revenue growth.
